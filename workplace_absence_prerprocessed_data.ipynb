# Import Statement

```python
import pandas as pd
```

# Data Cleaning

```python
raw_data = pd.read_csv("Absenteeism_data.csv")
```

```python
data = raw_data.copy()
```

```python
data.info()
```

```python
data = data.drop(['ID'], axis = 1)
```

# Data Preprocessing

### Modify the Reasons for absence column

The reason categories are health related and include:
- Reason_1: 1-14 = Diseases
- Reason_2: 15-17 = Pregnancy and childbirth
- Reason_3: 18-21 = Accidents and Poisonings
- Reason_4: 22-28 = Medical appointments

```python
data['Reason for Absence'].unique()
```

```python
sorted(data['Reason for Absence'].unique())
```

```python
reason_columns = pd.get_dummies(data['Reason for Absence'], drop_first = True)
```

```python
reason_columns
```

```python
reason_cat_1 = reason_columns.loc[:, 1:14].max(axis=1)
reason_cat_2 = reason_columns.loc[:, 15:17].max(axis=1)
reason_cat_3 = reason_columns.loc[:, 18:21].max(axis=1)
reason_cat_4 = reason_columns.loc[:, 22:].max(axis=1)
```

```python
data = data.drop(['Reason for Absence'], axis = 1)
```

```python
data = pd.concat([data, reason_cat_1, reason_cat_2, reason_cat_3, reason_cat_4], axis = 1)
data
```

```python
data.columns.values
```

```python
column_names = ['Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours', 'Reason_1', 'Reason_2', 'Reason_3', 'Reason_4']
```

```python
data.columns = column_names
```

```python
data.head()
```

```python
column_reorder = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4',
                          'Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']
```

```python
data = data[column_reorder]
```

```python
data.head()
```

```python
# @title Checkpoint
data_reason_mod = data.copy()
```

### Modify the Date column

```python
type(data_reason_mod['Date'][0])
```

```python
data_reason_mod['Date'] = pd.to_datetime(data_reason_mod['Date'], format = '%d/%m/%Y')
```

```python
data_reason_mod.info()
```

#### Extract Month

```python
list_months = []
```

```python
for i in range(data_reason_mod.shape[0]):
    list_months.append(data_reason_mod['Date'][i].month)
```

```python
data_reason_mod['Month'] = list_months
```

```python
data_reason_mod.head(20)
```

#### Extract Day

```python
def date_to_weekday(date_value):
    return date_value.weekday()
```

```python
data_reason_mod['Weekday'] = data_reason_mod['Date'].apply(date_to_weekday)
```

```python
data_reason_mod.head()
```

```python
data_reason_mod = data_reason_mod.drop(['Date'], axis = 1)
data_reason_mod.head()
```

```python
new_column_names = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4', 'Month', 'Weekday',
       'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education', 'Children',
       'Pets', 'Absenteeism Time in Hours']
```

```python
data_reason_mod = data_reason_mod[new_column_names]
data_reason_mod.head()
```

```python
# @title Checkpoint 2
data_reason_date_mod = data_reason_mod.copy()
data_reason_date_mod
```

### Modify the Education column

```python
data_reason_date_mod['Education'].value_counts()
```

```python
data_reason_date_mod['Education'] = data_reason_date_mod['Education'].map({1:0, 2:1, 3:1, 4:1})
```

```python
data_reason_date_mod['Education'].value_counts()
```

### Export *Data*

```python
# @title Checkpoint 3
df_preprocessed = data_reason_date_mod.copy()
df_preprocessed.head(10)
df_preprocessed.to_csv('Absenteeism_preprocessed.csv', index = False)
```
